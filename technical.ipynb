```python
def behaviour_score(s):
    st=((350*s["behaviour"])+(300*s["personality"])+(250*s["hobby"])+(200*s["hstudy"])+(150*s["adaptablilty"])+(100*s["decison"])+(50*s["life"]))
    if(s["match"]==0):
        st*=-1  
    return st
```
